# Color Function

```python
plt.figure(figsize=(16,9))
sns.heatmap(df,vmin=0,vmax=21,cmap='turbo')
plt.show()
```

# annot Function

```python
plt.figure(figsize=(16,9))
sns.heatmap(df,vmin=0,vmax=21,cmap='turbo',annot=True)
plt.show()
```

```python

annot_arr = np.array([['a00','a01','a02'],
                      ['a10','a11','a12'],
                      ['a20','a21','a22'],
                      ['a30','a31','a32']], dtype = str)
```

```python
arr_2d=np.linspace(1,5,12).reshape(4,3)
```

# fmt Function

```python
plt.figure(figsize=(16,9))
sns.heatmap(arr_2d, annot = annot_arr, fmt="s",cmap='coolwarm')
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(df,vmin=0,vmax=21,cmap='turbo',annot=True)
plt.show()
```

# annot_kws Function

```python
plt.figure(figsize=(16,9))

annot_kws=({'fontsize':12,
           'color':'k',
           'alpha':0.9,
            'rotation':'horizontal',
            'verticalalignment':'center',
            'backgroundcolor':'w'
           })
sns.heatmap(df,vmin=0,vmax=21,cmap='turbo',annot=True,annot_kws=annot_kws)
plt.show()
```

# linewidth Function

```python
plt.figure(figsize=(16,9))

annot_kws=({'fontsize':12,
           'color':'k',
           'alpha':0.9,
            'rotation':'horizontal',
            'verticalalignment':'center',
            'backgroundcolor':'w'
           })
sns.heatmap(df,vmin=0,vmax=21,cmap='coolwarm',annot=True,annot_kws=annot_kws,linewidth=4)
plt.show()
```

# linecolor Function

```python
plt.figure(figsize=(16,9))

annot_kws=({'fontsize':12,
           'color':'k',
           'alpha':0.9,
            'rotation':'horizontal',
            'verticalalignment':'center',
            'backgroundcolor':'w'
           })
sns.heatmap(df,vmin=0,vmax=21,cmap='coolwarm',annot=True,annot_kws=annot_kws,linewidth=4,linecolor='y')
plt.show()
```

# cbar Function Hide

```python
plt.figure(figsize=(16,9))

annot_kws=({'fontsize':12,
           'color':'k',
           'alpha':0.9,
            'rotation':'horizontal',
            'verticalalignment':'center',
            'backgroundcolor':'w'
           })
sns.heatmap(df,vmin=0,vmax=21,cmap='coolwarm',annot=True,annot_kws=annot_kws,linewidth=2,linecolor='y',
           cbar=False)
plt.show()
```

# xticklabels / yticklabels Function

```python
plt.figure(figsize=(16,9))

annot_kws=({'fontsize':12,
           'color':'k',
           'alpha':0.9,
            'rotation':'horizontal',
            'verticalalignment':'center',
            'backgroundcolor':'w'
           })
sns.heatmap(df,vmin=0,vmax=21,cmap='coolwarm',annot=True,annot_kws=annot_kws,linewidth=2,linecolor='y',
           cbar=False,xticklabels=False,yticklabels=False)
plt.show() 
```

```python
plt.figure(figsize=(25,20))
annot_kws=({'fontsize':12,
           'color':'k',
           'alpha':0.9,
            'rotation':'horizontal',
            'verticalalignment':'center',
            'backgroundcolor':'w'
           })


cbar_kws=({'orientation':'horizontal',
          'shrink':1,
           'extend':'min',
          'extendfrac':0.1,
          'ticks':np.arange(0,22),
          'drawedges':True})

sns.heatmap(df,vmin=0,vmax=21,cmap='coolwarm',annot=True,annot_kws=annot_kws,linewidth=2,linecolor='y',
          cbar_kws=cbar_kws)
plt.show() 
```
