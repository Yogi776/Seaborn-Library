# line Plot

```python
from numpy.random import randn
import pandas as pd
```

```python
df = pd.DataFrame(dict(time = np.arange(500), value = randn(500).cumsum()))
```

```python
df.head()
```

```python
df.describe().T
```

```python
df.info()
```

```python
sns.relplot(x='time',y='value',kind='line',data=df, sort=True, markers=True)
```

```python
fmri = sns.load_dataset('fmri')
fmri.head()
```

```python
sns.relplot(x='timepoint',y='signal',data=fmri,kind='line')
```

```python
fmri.columns
```

```python
fmri.event.unique()
```

```python
sns.relplot(x='timepoint',y='signal',data=fmri,kind='line',hue='event')
```

```python
sns.relplot(x = 'timepoint', y = 'signal', hue = 'region', kind = 'line', data = fmri,marker=True)
```

```python
sns.relplot(x = 'timepoint', y = 'signal', hue = 'region',  kind = 'line',style = 'event', data = fmri,marker=True)
```

```python
fmri.columns
```

```python
sns.relplot(x = 'timepoint', y = 'signal', hue = 'region',style = 'subject', data = fmri,marker=True,height=5)
```

```python
sns.relplot(x = 'timepoint', y = 'signal', hue = 'region',style = 'event', kind = 'line',  data = fmri,marker=True,dashes=False)
```

```python
sns.relplot(x = 'timepoint', y = 'signal', hue = 'region', style = 'event', kind = 'line', data = fmri, markers = True, dashes = False)
```

```python
sns.relplot(x = 'timepoint', y = 'signal', hue = 'region', style = 'event', kind = 'line', data = fmri, 
            markers = True, dashes = False,ci = 68)
```

```python
sns.relplot(x = 'timepoint', y = 'signal', hue = 'region', style = 'event', kind = 'line', data = fmri, 
            markers = True, dashes = False,ci = 68,err_style='bars')
```
